```python
import os
import zipfile
from glob import glob
from pathlib import Path
import numpy as np
import pandas as pd
import xarray as xr
import rioxarray
```

```python
start_date = '2022-03-31' 
end_date = '2024-09-05'


step = 60
periods_ = np.arange(np.datetime64(start_date),np.datetime64(end_date)-np.timedelta64(1),np.timedelta64(step))
periods_[-1] = np.datetime64(end_date) - np.timedelta64(1)
periods_
```

```python
fout = '/mnt/geodata/nsidc_sm/teste/out/'
Path(fout).mkdir( parents = True, exist_ok = True)
fin = '/mnt/geodata/nsidc_sm/teste/in/'
Path(fin).mkdir( parents = True, exist_ok = True)


bbox = '-74,-34,-29,6'
product = 'SPL3SMP_E'
version = '006'
b0 = 'curl -b ~/.urs_cookies -c ~/.urs_cookies -L -n -O -J --dump-header response-header.txt '
```

```python
for n in range(len(periods_)-1):
    print(n,)
    d1 = periods_[n]
    d2 = periods_[n+1]
    b1 = f"https://n5eil02u.ecs.nsidc.org/egi/request?short_name={product}&version={version}&format=GeoTIFF&time={d1},{d2}&bounding_box={bbox}&bbox={bbox}&Coverage=/Soil_Moisture_Retrieval_Data_AM/soil_moisture&projection=Geographic&page_size=100";
    print(n,b1)
    req = f'{b0}"{b1}"'
    os.chdir(fin)
    print(f'getting {d1} to {d2}')
    os.system(req)
    
    f = glob(f'{fin}*.zip')

    with zipfile.ZipFile(f[0], 'r') as zip_ref:
        zip_ref.extractall(fout)
    
    os.chdir(fout)
    os.system("find . -mindepth 2 -type f -print -exec mv {} . \;")
    os.system("find . -empty -type d -delete")
    os.system(f"rm {f[0]}")
```

```python
files = sorted(glob(fout+'*.tif'))
dates = pd.to_datetime([x.split('_')[6] for x in files])
geotiffs_da = xr.concat([rioxarray.open_rasterio(i, parse_coordinates=True) for i in files], dim=dates)
geotiffs_da = xr.where(geotiffs_da == -9.9990000e+03, np.nan, geotiffs_da)
smwr = geotiffs_da.resample(concat_dim = '1W').max()
smwr = smwr.rio.write_crs("epsg:4326")
smwr = smwr.rename({'x': 'longitude','y': 'latitude', 'concat_dim' : 'time'})
smwr.to_netcdf(fout+f'_SM_week_v0_{str(periods_[0])}-{str(periods_[-1])}.nc')
```
